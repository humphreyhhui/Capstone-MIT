## **Problem Definition**

### **The Context:**

 - Why is this problem important to solve?

In today’s competitive market, understanding customers is essential for strategic growth. Consumers have diverse preferences, behaviors, and decision-making processes. Ignoring these differences leads to generic experiences that fail to engage customers, resulting in missed opportunities, lower satisfaction, and reduced revenue.

For a company with a broad product portfolio, optimizing the customer experience is critical. Knowing who your customers are, what they value, and how they behave allows you to streamline the sales process and anticipate their needs. Effective customer engagement doesn’t just improve satisfaction — it drives better business outcomes. When you target the right customers with the right strategies and channels, you increase efficiency, strengthen loyalty, and achieve more profitable growth.

### **The objective:**

 - What is the intended goal?
   
Our goal is to segment the customer base using behavioral and demographic data to create a framework for personalized engagement. This segmentation will help us allocate resources effectively to maximize impact.

For instance, if we identify a group that responds well to discounts, we can design targeted promotions to boost conversions. For segments less motivated by price, we can create value-based offers, protecting margins.

Segmentation also enables us to optimize communication channels, ensuring customers receive messages through their preferred platforms, such as email, social media, or direct outreach. Additionally, tailored product recommendations can enhance cross-selling and up-selling opportunities, increasing customer lifetime value.

Ultimately, strategic segmentation improves customer satisfaction and business efficiency, positioning the company for sustainable growth.

Optimal Segmentation Models:

Business Strategy Alignment:
- How can these segments be used to improve marketing, sales, and customer service strategies?
- What KPIs (e.g., conversion rates, customer retention, average spend) will measure the effectiveness of these strategies?

### **Applying KMeans on the PCA data and visualizing the clusters**

```python
# Initializing a dictionary to store the Sum of Squared Error (SSE) for each K.
SSE = {}

# Running for a range of Ks and store SSE for each run.
for k in range(2,10):
    kmeans = KMeans(n_clusters = k, random_state = 1).fit(scaled_df_behavioural_pca)
    SSE[k] = kmeans.inertia_

# Plotting the SSE vs K and plotting the elbow curve.
plt.plot(list(SSE.keys()), list(SSE.values()))
plt.title('SSE against Number of Clusters')
plt.xlabel("Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
# Slight elbow at 3. Checking silhouette scores.
silhouette_scores_kmeans = {} 
for k in range(2, 10):
    kmeans = KMeans(n_clusters = k, random_state = 1).fit(scaled_df_behavioural_pca)
    labels = kmeans.labels_
    silhouette_scores_kmeans[k] = silhouette_score(scaled_df_behavioural_pca, labels)

print(silhouette_scores_kmeans)

plt.figure()
plt.plot(list(silhouette_scores_kmeans.keys()), list(silhouette_scores_kmeans.values()))
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette score")
plt.show()
```

```python
# 3 has a higher silhouette score. Therefore, 3 should be chosen as the value of K.
kmeans = KMeans(n_clusters = 3, random_state = 1)
kmeans.fit(scaled_df_behavioural_pca)

# Adding K-Means label to all dfs.
scaled_df_behavioural_pca['KMeans_Labels'] = kmeans.labels_
df_behavioural['KMeans_Labels'] = kmeans.labels_
df['KMeans_Labels'] = kmeans.labels_

# Checking how many are in each group.
df_behavioural['KMeans_Labels'].value_counts()
```

## **DBSCAN**

- Although DBSCAN had the highest silhouette score, it was only able to identify one cluster instead of multiple.

```python
# Set potential EPS and min_sample variables to loop through.
eps_values = [3, 5, 7, 10]
min_sample_values = [5, 10, 15]
highest_silhouette_avg = -1
best_eps = None
best_min_samples = None

# Create DBSCAN for loop to go through EPS values and min_sample values to find highest silhouette score.
for eps in eps_values:
    for min_samples in min_sample_values:
        db = DBSCAN(eps = eps, min_samples = min_samples).fit(scaled_df_behavioural_pca)
        labels = db.labels_
        if len(set(labels)) > 1:
            silhouette_avg = silhouette_score(scaled_df_behavioural_pca, labels)
            if silhouette_avg > highest_silhouette_avg:
                highest_silhouette_avg = silhouette_avg
                best_eps = eps
                best_min_samples = min_samples

print(f"Highest silhouette average is {highest_silhouette_avg:.2f} for eps = {best_eps} and min sample = {best_min_samples}")
```
